```python
# Load datasets

apps_df = pd.read_csv("Play Store Data.csv")
reviews_df = pd.read_csv("User Reviews.csv")
```

```python
# Check dataset shapes

print("Play Store Data Shape:", apps_df.shape)
print("User Reviews Data Shape:", reviews_df.shape)
```

```python
# Display first 5 rows

apps_df.head()
```

```python
reviews_df.head()
```

```python
# Display column names

print("Play Store Columns:")
print(apps_df.columns)

print("\nReviews Dataset Columns:")
print(reviews_df.columns)
```

```python
# Check missing values

print("Missing values in Play Store Dataset:")
print(apps_df.isnull().sum())

print("\nMissing values in Reviews Dataset:")
print(reviews_df.isnull().sum())
```

```python
# Remove rows with missing ratings

apps_df = apps_df.dropna(subset=['Rating'])
```

```python
# Remove duplicate apps

apps_df = apps_df.drop_duplicates(subset='App')
```

```python
# Remove rows where review text is missing

reviews_df = reviews_df.dropna(subset=['Translated_Review'])
```

```python
# Convert Reviews column to numeric

apps_df['Reviews'] = pd.to_numeric(apps_df['Reviews'], errors='coerce')
```

```python
# Check dataset size after cleaning

print("Play Store Data Shape after cleaning:", apps_df.shape)
print("Reviews Data Shape after cleaning:", reviews_df.shape)
```

```python
# Clean Installs column safely

apps_df['Installs'] = apps_df['Installs'].astype(str)

apps_df['Installs'] = apps_df['Installs'].str.replace(',', '', regex=False)
apps_df['Installs'] = apps_df['Installs'].str.replace('+', '', regex=False)

apps_df['Installs'] = pd.to_numeric(apps_df['Installs'], errors='coerce')

apps_df = apps_df.dropna(subset=['Installs'])

apps_df['Installs'] = apps_df['Installs'].astype(int)
```

```python
apps_df[['Installs']].head()
```

```python
# Clean Price column

apps_df['Price'] = apps_df['Price'].str.replace('$', '')
apps_df['Price'] = apps_df['Price'].astype(float)
```

```python
# Function to convert size to MB

def convert_size(size):
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024
    else:
        return None

apps_df['Size'] = apps_df['Size'].astype(str)
apps_df['Size_MB'] = apps_df['Size'].apply(convert_size)
```

```python
# Convert Last Updated to datetime

apps_df['Last Updated'] = pd.to_datetime(apps_df['Last Updated'], errors='coerce')
```

```python
# Extract month and year

apps_df['Update_Month'] = apps_df['Last Updated'].dt.month
apps_df['Update_Year'] = apps_df['Last Updated'].dt.year
```

```python
# Extract Android version number

apps_df['Android_Version'] = apps_df['Android Ver'].str.extract(r'(\d+\.\d+)')
apps_df['Android_Version'] = pd.to_numeric(apps_df['Android_Version'], errors='coerce')
```

```python
# Display transformed columns

apps_df[['Installs', 'Price', 'Size_MB', 'Update_Month', 'Update_Year', 'Android_Version']].head()
```

```python
# Aggregate sentiment data by App

reviews_agg = reviews_df.groupby('App').agg({
    'Sentiment_Polarity': 'mean',
    'Sentiment_Subjectivity': 'mean'
}).reset_index()
```

```python
# Merge Play Store data with aggregated review data

merged_df = pd.merge(apps_df, reviews_agg, on='App', how='left')
```

```python
# Check merged dataset

merged_df.head()
```

```python
# Check missing sentiment values

merged_df[['Sentiment_Polarity','Sentiment_Subjectivity']].isnull().sum()
```

```python
# Fill missing sentiment values

merged_df['Sentiment_Polarity'] = merged_df['Sentiment_Polarity'].fillna(0)
merged_df['Sentiment_Subjectivity'] = merged_df['Sentiment_Subjectivity'].fillna(0)
```

```python
merged_df
```

```python
# Create Revenue column

merged_df['Revenue'] = merged_df['Price'] * merged_df['Installs']
```

```python
# Calculate app name length

merged_df['App_Name_Length'] = merged_df['App'].apply(len)
```

```python
# Detect if app name contains numbers

merged_df['Has_Number'] = merged_df['App'].str.contains(r'\d', regex=True)
```

```python
import plotly.express as px
import plotly.io as pio

# Step 1: Copy dataset
task2_df = merged_df.copy()
```

```python
# Step 2: Find top 5 categories by installs
top_categories = (
    task2_df.groupby('Category')['Installs']
    .sum()
    .nlargest(5)
    .index
)

task2_df = task2_df[task2_df['Category'].isin(top_categories)]
```

```python
# Step 3: Group installs by category
category_installs = (
    task2_df.groupby('Category')['Installs']
    .sum()
    .reset_index()
)
```

```python
# Step 4: Create dummy countries for choropleth map
country_mapping = {
    category_installs['Category'].iloc[0]: ('USA', 'USA'),
    category_installs['Category'].iloc[1]: ('India', 'IND'),
    category_installs['Category'].iloc[2]: ('Germany', 'DEU'),
    category_installs['Category'].iloc[3]: ('Brazil', 'BRA'),
    category_installs['Category'].iloc[4]: ('Canada', 'CAN')
}

category_installs['Country'] = category_installs['Category'].map(
    lambda x: country_mapping[x][0]
)

category_installs['ISO_Code'] = category_installs['Category'].map(
    lambda x: country_mapping[x][1]
)
```

```python
# Step 5: Highlight condition
category_installs['Highlight'] = category_installs['Installs'] > 1000000
```

```python
# Step 6: Create Choropleth Map
fig_task2 = px.choropleth(
    category_installs,
    locations='ISO_Code',
    locationmode='ISO-3',
    color='Installs',
    hover_name='Category',
    hover_data={
        'Country': True,
        'ISO_Code': False,
        'Installs': True,
        'Highlight': True
    },
    color_continuous_scale='Blues',
    title='Global Installs by Category'
)
```

```python
# Step 7: Improve border visibility
fig_task2.update_traces(
    marker_line_color='black',
    marker_line_width=1
)
```

```python
# Step 8: Improve layout
fig_task2.update_layout(
    template='plotly_white',
    title_x=0.5,
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='equirectangular'
    )
)

fig_task2.show()
```

```python
# Step 9: Save chart
pio.write_html(fig_task2, "task_2_global_installs_by_category.html", full_html=True)

print("Task 2 completed and saved as HTML")
```
